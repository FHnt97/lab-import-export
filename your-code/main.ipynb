```python
film= pd.read_sql(query3, engine)
film.head()
```

```python
actor_film_names = actor_film.merge(actor, how= "inner", on = "actor_id")
actor_film_names.head()
```

```python
query4 = """ SELECT title, release_year, f.film_id, first_name AS actor_name, last_name AS actor_2_name
FROM actor as a
LEFT JOIN film_actor as af
ON a.actor_id = af.actor_id

LEFT JOIN film as f
ON af.film_id = f.film_id



"""
new_af= pd.read_sql(query4, engine)
new_af.head()
```
